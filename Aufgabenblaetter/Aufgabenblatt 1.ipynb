# Aufgabe 1.14

```python
from matplotlib import pyplot as plt
from collections import Counter
from math import ceil
%matplotlib inline
#a
def collatz(i, max_iter=0, plot=False):
    if i == 0 : return None
    l = []
    if max_iter != 0:
        for j in range(max_iter):
            if i % 2 == 0:
                candidate = i/2
            else:
                candidate = 3*i+1
            l.append(candidate)
            if candidate == 1 : break
            i = candidate
    else:
        while(True):
            if i % 2 == 0:
                candidate = i/2
            else:
                candidate = 3*i+1
            l.append(candidate)
            if candidate == 1 : break
            i = candidate
    if plot:
        lengths = len(l)
        average = sum(l)/lengths
        plt.plot(range(lengths)[:], l)
        plt.plot(range(lengths)[:],[average]*lengths, "r-", label = "avg at {}".format(average))
        plt.legend()
        plt.ylabel("Wert bei Index i")
        plt.xlabel("Index i")
        plt.title("Collatz-Folge")
    return l

#b
def lenCollatz(n):
    return len(collatz(n))

#c
_ = collatz(100000, plot=True)

#d - for execution see cell below
def lenColsFrq(n, num_bins=10, h=-1):
    lengths = [ lenCollatz(_) for _ in range(1,n) if lenCollatz(_) > h]
    plt.hist(lengths, bins=num_bins)
    plt.title("Verteilung von Collatz-Folgen Laengen")
    plt.xlabel("Laenge der Collatz-Folge")
    plt.ylabel("Haeufigkeit der Laenge")
    
#e
#siehe #d
```

# Aufgabe 1.16*

```python
def perm(n):
    result = [] # collecting all results
    if(len(n) == 1):
        result = n[0]
        return result 
    iterations = len(n)
    for i in range(iterations):
        number = n[i] # pick a number
        n_copy = n[:] # clone original list to not mess up the initial source list
        n_copy.remove(number) # "subtract" the number from the list
        if len(n_copy) > 1: # if there is more than just one element left
            combinations = [ [number] + x for x in perm(n_copy) ] # Inception, add number in front of every combination
            result.append(combinations)
        else: # list has only 2 elements 
            result.append([number, n_copy[0]])
    if iterations > 2:
        result = reduce(lambda x,y : x+y, result)# flatten out the results since they are nested per iteration
    return result
perm([1,2,3])
```

# Aufgabe 1.17*

```python
def choose(L,k):
    #checks to avoid recustion
    if k > len(L) or k < 1 : return None # not defined
    if k == 1 : return [ [number] for number in L ] # simply split the array
    result = []
    if(len(L) == 1):
        result.append(L[0])
        return result
    iterations = len(L)-k+1 # postition of last iteration 
    for _ in range(iterations):
        number = L[_]
        L_copy = L[_+1:] # everything after the current index
        if k == 2:
            result.append([ [number] + [x] for x in L_copy ])
        else:
            result.append([ [number] + x for x in choose(L_copy, k-1) ]) # rec
    
    result = (reduce(lambda x,y: x+y, result)) # flatten nested arrays
    return result
choose([1,2,3,4],2)
```

```python
C = 'abcdefghijklmnopqrstuvwxyz'
D = ''.join(map(str, range(0,10)))
A = C+D
# a
def woerter(n):
    if n == 0 : return None
    if n == 1: return [ char for char in A ]
    result = []
    for i in range(n):
            result = [ x + y for x in woerter(1) for y in woerter(n-1) ]
    return result
print("A\n{}\n".format(woerter(2))) # ugly for woerter n if n > 1 ...
        
# b - Q.A.D. :)
def w1d(n):
    return [ item for item in woerter(n) if not item.isalpha()  ]
print("B\n{}\n".format(w1d(2)))

# c - Q.A.D. :)
def wdE(n):
    return [ item for item in woerter(n) if item[-1].isdigit() ]
print("C\n{}\n".format(wdE(2)))
```

# Aufgabe 1.19

```python
# in Anlehnung and das Map-Reduce-Paradigma
import re # regular expressions
from collections import defaultdict

def mapWortDateiPosition(lst):
    result = []
    [ result.append([ wort, f[0], position.start()]) \
     for f in lst \
     for wort in set(f[1].split()) \
     for position in re.finditer(wort, f[1])]
    return result

def reduceByWort(lst):
    result = defaultdict(list) # defaultdict(list) will append and handle if key not already existing
    [ result[wort_triple[0]].append(wort_triple[1:]) for wort_triple in lst ]
    return result

def reduceByDatei(lst):
    result = {}
    tmp = defaultdict(list)
    for key in lst.keys():
        for item in lst.get(key):
            tmp[item[0]].append(item[1])
        result[key] = tmp.copy()
        tmp.clear()
    return result

def pretty_print(lst):
    keys = lst.keys()
    keys.sort(key=lambda x: x.upper()) # sort ignoring upper and lower case
    for key in keys:
        print("\n{}".format(key) + "\n " + u"\u2533")
        keys2 = lst[key].keys()
        keys2.sort(key=lambda x: x.upper()) # sort ignoring upper and lower case
        len_keys = len(keys2)
        i = 1
        for key2 in keys2:
            if i != len_keys:
                prefix = u"\u2523"
            else:
                prefix = u"\u2517"
            print(" " + prefix + " {} at index:{}".format(key2, lst[key].get(key2)))
            i += 1

# Analyze some Star Wars Data :)
files = [ f for f in listdir('.') if f.startswith("StarWars") ]
starwars_raw = [ (f, open(f).read()) for f in files ]

# start the analysis (1 mapping step and 2 reduce steps )
mWDP = mapWortDateiPosition(starwars_raw) # Map 
rbW = reduceByWort(mWDP) # Reduce by Wort
rbD = reduceByDatei(rbW) # Reduce by Datei

# Pretty Print
print("RESULT:\n*******")
pretty_print(rbD)
    
# Print the steps to produce the result
# Map
print("\n")
print("MAPPING:\n********")
for item in mWDP:
    print(item)
print("\n")
# Reduce by Wort
print("REDUCE BY WORT:\n***************")
for key in rbW.keys():
    print("{} : {}".format(key, rbW.get(key)))
print("\n")
# Reduce by Datei
print("REDUCE BY DATEI:\n****************")
for key in rbD.keys():
    print("{} : {}".format(key, str(rbD.get(key))[27:-1])) # ugly but removes unwanted prefix
print("\n")
```
